```python
%matplotlib inline

import numexpr as ne
import matplotlib.pyplot as plt
import numpy as np

import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output

f_input = widgets.Text(description="Function: ")
button  = widgets.Button(description="Evaluate")
noise = widgets.IntSlider(value=20, min=0, max=100, description="Noise: ")
output  = widgets.Output()
graph_out = widgets.Output()

x_range = range(100)

# Interactive mode
plt.ioff()
ax = plt.axes()
def plot_function(btn):
    gaussian_noise = noise.value * np.random.normal(0, 1, size=len(x_range))
    
    values = []
    for x in x_range:
        value = ne.evaluate(f_input.value)
        values.append(value)

    noised_values = values + gaussian_noise
    
    ax.plot(x_range, noised_values, 'x')
#     with graph_out:
#         clear_output(wait=True)
#         display(ax.figure)

display(f_input, noise, button, output)
button.on_click(plot_function)
```

```python
import tensorflow as tf
import numexpr as ne
import numpy as np 
import matplotlib.pyplot as plt
```

```python
sampling_freq = 10
n = int(sampling_freq * 4*np.pi)

x = np.linspace(-2*np.pi,2*np.pi, n)
noise_level = 1
epsilon = [noise_level * np.random.rand() if np.random.rand() < 0.5 else noise_level * -1 * np.random.rand() for _ in range(n)]

func = "3*sin(2*x) + 0.5*x**2"
signal_freq = 1 / (4*np.pi)
shannon_limit = (2*signal_freq)

X_train = x
y_clean = ne.evaluate(func)
y_train = y_clean + epsilon
plt.scatter(X_train, y_train)

plt.plot(X_train, y_clean)

max_num = np.max(y_train)
min_num = np.min(y_train)

print(f"Signal frequency: {signal_freq}, sampling frequency: {sampling_freq},\n Shannon sampling rate: {shannon_limit}")
```

```python
n_neurons = 200
n_epochs = 50

model = tf.keras.Sequential()


opt = tf.keras.optimizers.Adam(learning_rate=0.1)

model.add(tf.keras.layers.Dense(n_neurons, activation=tf.keras.activations.sigmoid))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer=opt, loss='mse')

# plt.ioff()
# ax = plt.axes()
fig = plt.figure()
figures = []

for epoch in range(n_epochs):
    model.fit(X_train, y_train, epochs=1)

    X_test = np.linspace(-2*np.pi, 2*np.pi, 200)
    preds = model.predict(X_test)
    plt.plot(X_test, preds)
    plt.plot(X_train, y_train, 'o')
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(min_num, max_num)
    
    figures.append(plt.figure(1))
    plt.close()
```

```python
import imageio

folder_img = 'images/'
folder_gif = "gifs/"

gif_file = folder_gif + f'sin_approx-{n_neurons}.gif'

with imageio.get_writer(gif_file, mode='I') as w:
    for n in range(n_epochs):
        image_file = folder_img + f'{n}.png'
        figures[n].savefig(image_file)
        image = imageio.imread(image_file)
        w.append_data(image)

from IPython.display import Image
Image(open(gif_file, 'rb').read())
```
